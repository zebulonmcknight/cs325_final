```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, log_loss, confusion_matrix
```

```python
X_df = pd.read_csv("data/processed_samples.csv")
y_df = pd.read_csv("data/targets.csv")
```

```python
X = X_df.to_numpy().astype(float)
y = y_df.to_numpy().astype(float)
```

```python
X_train, y_train, X_val, y_val = train_test_split(X, y, test_size = 0.2, stratify=y)
```

```python
logreg = LogisticRegression(
    multi_class='multinomial',  # for true multiclass
    solver='lbfgs',             # good default solver
    max_iter=1000,              # higher to ensure convergence
    C=1.0,                      # regularization strength
    class_weight='balanced',    # optional, for imbalanced classes
    random_state=42             # for reproducibility
)
```

```python
logreg.fit(X_train, y_train)
```
